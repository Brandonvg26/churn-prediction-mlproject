## Batch Inference with Automatic Feature Retrieval
This step demonstrates the value of the FeatureLookup pattern. At inference time, you only need to provide the primary key (customer_id). The model retrieves all features from the feature store automatically — no manual feature joins required.

```python
import mlflow

# Load the registered model from Unity Catalog
model_uri = 'models:/ml1.ml_project.churn_prediction_model/1'


# Inference input: only the lookup key
# All features are retrieved automatically from the feature store
inference_df = spark.table('ml1.ml_project.churn_labels') \
    .select('customer_id')


# score_batch handles the feature retrieval transparently
predictions_df = fe.score_batch(
    model_uri=model_uri,
    df=inference_df
)

predictions_df.select('customer_id', 'prediction').show(10)


# Write predictions to a results table
predictions_df.write.format('delta') \
    .mode('overwrite') \
    .saveAsTable('ml1.ml_project.churn_predictions')
```
